# L4 GPU: Alexnet vs ResNet18

```python
import matplotlib.pyplot as plt
import numpy as np

total_flops_list = [10486620979200, 23213781811200]        
total_cuda_time_list = [0.774669, 1.360620]  # Total CUDA time in seconds
measured_performance_list = [13536.91, 17061.18]  # Measured performance in GFLOPS/s

# GPU peak performance (121 TFLOPS)
peak_performance = 121 * 1000  # 121000 GFLOPS/s

# Memory bandwidth (in GB/s)
memory_bandwidth = 300

titles = ["AlexNet", "ResNet18"]
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Define  range of arithmetic intensities 
intensity_range = np.logspace(-1, 3, 400)  # from 0.1 to 1000 FLOPs/Byte
memory_bound_line = memory_bandwidth * intensity_range

for i in range(2):
    measured_performance = measured_performance_list[i]
    # Calculate measured arithmetic intensity
    measured_intensity = measured_performance / memory_bandwidth
    
    ax = axes[i]
    # Plot the memory bound line
    ax.loglog(intensity_range, memory_bound_line, label=f'Memory Roof (BW = {memory_bandwidth} GB/s)', lw=2)
    # Plot the compute roof as a horizontal line
    ax.axhline(peak_performance, color='r', linestyle='--', label=f'Compute Roof (Peak = {peak_performance} GFLOPS/s)', lw=2)
    # Plot the measured performance point
    ax.scatter(measured_intensity, measured_performance, color='k', s=100, zorder=5, label='Measured Performance')
    ax.text(measured_intensity * 1.2, measured_performance * 0.8,
            f'({measured_intensity:.2f}, {measured_performance:.2f})', fontsize=10)
    
    ax.set_xlabel('Arithmetic Intensity (FLOPs/Byte)', fontsize=12)
    ax.set_ylabel('Performance (GFLOPS/s)', fontsize=12)
    ax.set_title(titles[i], fontsize=14)
    ax.legend()
    ax.grid(True, which="both", ls="--", linewidth=0.5)

plt.tight_layout()
plt.show()
```
